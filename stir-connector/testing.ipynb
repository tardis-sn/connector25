With the current setup and values in the config.yaml file, to use this you only need to do the following:

* Download the EOS file. It should be SFHo.h5 and is included in the STIR folder on the tardis connector onedrive. I put it in the main folder, but you can specify it's location in the config file.
* Place the .mod and .data MESA progenitor files in the progenitors folder. Both need to have the same name, and this name is what you use as the first argument in the convert method below.
* Place the STIR profile in the stir_profiles folder. It's name should be the model name, followed by "_a1.25" where 1.25 is your alpha. This alpha is the second argument in the convert method.
* Specify the stir_portion in the convert method. This is the inner percentage of the STIR domain (with radius) that is kept, allowing us to cut out garbage near the boundary. 0.9 is a good place to start. After conversion, look at the profiles. If there is a discontinuity at the intersection, you may want to lower it.
* If the plots show no or very little blue region, check the total specific energy plot. The energy may never go above 0 in the STIR domain.
